```python
import gensim
import re
import nltk
from nltk.corpus import stopwords           #Used to remove common words like "is", "the", "on" etc.
from nltk.stem import WordNetLemmatizer     #Converts words to their base form (e.g., "playing" â†’ "play").
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess

# Download NLTK data (run once)
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

## Preprocessing Corpus

```python
# Initialize
stop_words = set(stopwords.words('english')) # It loads english stopwords into a sset for fast lookup
lemmatizer = WordNetLemmatizer()

sentences = [
    "I love natural language processing",
    "Word2Vec creates vector representations of words",
    "Gensim makes working with word embeddings easy",
    "Python is great for NLP tasks"
]
# Preprocessing function
def preprocess_manual(sentence):
    # 1. Lowercasing
    sentence = sentence.lower()
    # 2. Remove punctuation using regex
    sentence = re.sub(r'[^\w\s]', '', sentence)

    # 3. Tokenize by splitting on whitespace
    tokens = sentence.split()
    
    # 4. Remove stopwords 
    tokens = [word for word in tokens if word not in stop_words]
    # Remove numbers
    sentence = re.sub(r'\d+', '', sentence)
    # Remove extra spaces
    sentence = sentence.strip()

    # 5. Lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens] # Converts word from
    
    return tokens

    # Apply preprocessing
tokenized_sentences = [preprocess_manual(sentence) for sentence in sentences] # Prepares list from the preprocessed sentences 

print("Preprocessed corpus:")
print(tokenized_sentences)
```

## Word2Vec Approach

```python
model_cbow = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=50,  # dimensionality of word vectors
    window=3,        # context window
    min_count=1,     # include all words
    sg=0             # 0 = CBOW, 1 = Skip-Gram
)
```

```python
# Vector for a word
print("Vector for 'python':\n", model_cbow.wv['python'])

# Most similar words
print("\nMost similar words to 'python':")
print(model_cbow.wv.most_similar('python'))

# Similarity between two words
print("\nSimilarity between 'python' and 'nlp':", model_cbow.wv.similarity('python', 'nlp'))
```

```python
# Save model
model_cbow.save("word2vec_cbow_preprocessed.model")

# Load model
loaded_model = Word2Vec.load("word2vec_cbow_preprocessed.model")
print("Model loaded successfully!")
```

## Skipgram Approach

```python
from gensim.models import Word2Vec

# Train Skip-Gram model
model_skipgram = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=50,  # size of embeddings
    window=3,        # context window
    min_count=1,     # include all words
    sg=1             # 1 = Skip-Gram
)
```

```python
# Vector for a word
print("Vector for 'python':\n", model_skipgram.wv['python'])

# Most similar words
print("\nMost similar words to 'python':")
print(model_skipgram.wv.most_similar('python'))

# Similarity between two words
print("\nSimilarity between 'python' and 'nlp':", model_skipgram.wv.similarity('python', 'nlp'))
```

```python
# Save model
model_skipgram.save("word2vec_skipgram_preprocessed.model")

# Load model
loaded_skipgram = Word2Vec.load("word2vec_skipgram_preprocessed.model")
print("Skip-Gram model loaded successfully!")
```
