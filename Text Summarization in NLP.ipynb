```python
# ------------------------------
# 1. Install required packages (run in notebook once)
# ------------------------------
# !pip install spacy pytextrank pandas
# !python -m spacy download en_core_web_sm

import pandas as pd
import spacy
import pytextrank

# ------------------------------
# 2. Load CSV File
# ------------------------------
df = pd.read_csv("television_reviews.csv")

# Auto-detect the column with reviews/comments/text
review_column = None
for col in df.columns:
    if "review" in col.lower() or "text" in col.lower() or "comment" in col.lower():
        review_column = col
        break

if review_column is None:
    raise ValueError("No column found with 'review', 'text', or 'comment' in the name!")

print(f" Using column: {review_column}")

# Combine all reviews into one large text (drop missing values)
text = " ".join(df[review_column].dropna().astype(str))
print("Original Document Size:", len(text))

# ------------------------------
# 3. Load spaCy + add TextRank (small model to avoid OSError)
# ------------------------------
nlp = spacy.load("en_core_web_sm")  # small English model (lg, md is not working as it is larger model ) 
nlp.add_pipe("textrank")            # add TextRank summarizer

# ------------------------------
# 4. Process the text
# ------------------------------
doc = nlp(text)

# ------------------------------
# 5. Extract key phrases
# ------------------------------
print("\n Top 10 Key Phrases:")
for phrase in doc._.phrases[:10]:
    print(f"- {phrase.text} (Rank: {phrase.rank:.4f}, Count: {phrase.count})")

# ------------------------------
# 6. Extract summary sentences
# ------------------------------
print("\n Extractive Summary (Top 5 sentences):")
for sent in doc._.textrank.summary(limit_phrases=5, limit_sentences=5):
    print("-", sent.text.strip())
```

## TextRank Algorithm (TextRank is a graph-based algorithm (like PageRank for web pages))

## Example Text (Input) "This phone has an excellent battery life and a very smooth display. However, 
## the camera quality is poor and the price feels too high. Overall, its good for performance but not for photography."

### answer: Phone has excellent battery life and smooth display. Camera quality is poor

```python
# -------------------------------
# Install dependencies (run once)
# -------------------------------
# !pip install spacy pytextrank
# !python -m spacy download en_core_web_sm

import spacy   # Help in Tokenization,Sentence splitting,POS tagging(Part-of-Speech,verb ,noun)
import pytextrank # the TextRank implementation that integrates

# Load English model
nlp = spacy.load("en_core_web_sm") # pre-trained model of nlp of tokenization, split text

# Add PyTextRank to pipeline
nlp.add_pipe("textrank") # Build a graph of phrases/sentences,Run the TextRank algorithm,

# Sample text
text = """
The iPhone 15 has an amazing camera that takes clear photos even in low light. 
The battery lasts almost two days, which is very impressive. 
However, the price is very high, making it less affordable for many people.
"""

doc = nlp(text)

print("=== Extractive Summary (TextRank) ===")
for sent in doc._.textrank.summary(limit_sentences=2):  # Top 2 important sentences
    print(sent.text)
```

```python
# -------------------------------
# Install dependencies (run once)
# -------------------------------
# !pip install spacy pytextrank pandas
# !python -m spacy download en_core_web_sm

import spacy
import pytextrank
import pandas as pd

# -------------------------------
# Load SpaCy model + PyTextRank
# -------------------------------
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("textrank")

# -------------------------------
# Load CSV file
# -------------------------------
df = pd.read_csv("television_reviews.csv")  # Replace with your file path

# Automatically pick the first text column (object type)
text_column = df.select_dtypes(include=['object']).columns[0]

# -------------------------------
# Function to summarize a single text
# -------------------------------
def summarize_text(text, limit=2):
    doc = nlp(str(text))
    summary = [sent.text for sent in doc._.textrank.summary(limit_sentences=limit)]
    return " ".join(summary)

# -------------------------------
# Apply summarization to each row
# -------------------------------
df_summary = pd.DataFrame()
df_summary["summary"] = df[text_column].apply(summarize_text)

# -------------------------------
# Save only summaries to CSV
# -------------------------------
df_summary.to_csv("television_summaries.csv", index=False)

print("Summarization complete! Sample output:")
print(df_summary.head())
```
